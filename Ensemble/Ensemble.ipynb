```python
x = df.drop(['RETURNED_2ND_YR'], axis=1)
y = df['RETURNED_2ND_YR']
```

```python
x = pd.get_dummies(x, drop_first=True)
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
col = x.columns

sc = MinMaxScaler()
x = sc.fit_transform(x)
x = pd.DataFrame(x, columns=col)
x.head()
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier

model = RandomForestClassifier()
model.fit(x, y)
imp = pd.DataFrame(model.feature_importances_, index=x.columns)
imp.nlargest(70, 0).plot(kind='barh', figsize=(10, 5))
```

```python
from sklearn.feature_selection import SelectKBest, chi2
best_features = SelectKBest(score_func=chi2, k=70)
# can not get a negative value so i did not use Standared Scaler
fit = best_features.fit_transform(x, y)
# select the features col
selected = x.columns[best_features.get_support()]
selected_x = x[selected]
selected_x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    selected_x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
```

```python
# models
models = {
    'LR': LogisticRegression(),
    'DT': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(),
    'SVC': SVC(),
    'NB': GaussianNB(),
    'RF': RandomForestClassifier(),
    'ExtraTree': ExtraTreesClassifier(),
    'Bagging': BaggingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'XGB': XGBClassifier()
}

# ensemble model
ensemble = {
    'RF': RandomForestClassifier(),
    'ExtraTree': ExtraTreesClassifier(),
    'Bagging': BaggingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'XGB': XGBClassifier()
}

# all parameters
params = {
    'LR': {'penalty': ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]},
    'DT': {'criterion': ['gini', 'entropy'], 'max_depth': [100, 200, 300]},
    'KNN': {'n_neighbors': [2, 3, 4]},
    'SVC': {'C': [0.001, 0.01, 0.1, 1, 10,], 'kernel': ['rbf', 'sigmoid']},
    'NB': {},

    'RF': {'n_estimators': [100, 200, 300]},
    'ExtraTree': {'n_estimators': [100, 200, 300]},
    'Bagging': {'n_estimators': [100, 200, 300]},
    'AdaBoost': {'n_estimators': [100, 200, 300]},
    'GradientBoosting': {'n_estimators': [100, 200, 300]},
    'XGB': {'n_estimators': [100, 200, 300], 'learning_rate': [0.01, 0.05, 0.1]}
}

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)

best_model = {}

for name, model in models.items():
    clf = GridSearchCV(model, params[name],
                       cv=cv, scoring='accuracy', n_jobs=-1)
    clf.fit(x_train_smote, y_train_smote)
    best_model[name] = clf.best_estimator_


from imblearn.over_sampling import SMOTE
from sklearn.ensemble import VotingClassifier

estimators = [('rf', best_model['RF']), ('XGBoost', best_model['XGB']),
              ('AdaBoost', best_model['AdaBoost']), ('NB', best_model['NB'])]
vote = VotingClassifier(estimators=estimators, voting='soft')
vote.fit(x_train_smote, y_train_smote)
```

```python
best_model['vote'] = vote
```

```python
mx = 0
mdl = ""
for name, model in best_model.items():
    y_pred = model.predict(x_test)
    print(f"{name} -> {accuracy_score(y_test,y_pred)}")
    print(classification_report(y_test, y_pred))
    print()

    if accuracy_score(y_test, y_pred) > mx:
        mx = accuracy_score(y_test, y_pred)
        mdl = name

print("------------------")
print(f"{mdl} -> {mx}")
```

```python
# plot cm for GradientBoosting
y_pred = best_model['GradientBoosting'].predict(x_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
```
